###1.DESIGN A PROCESS FOR A SINGLE URL

### Get Product Name & Brand

```python
boxes = soup.find_all('div',attrs={'class':'StyledCardWrapper-sc-1j5kbg6-0 jBJHbA'})

brands = ['LENOVO','HP','ASUS','APPLE','ACER','MICROSOFT','HUAWEI','MSI','DYNABOOK','SAMSUNG','TOSHIBA','INNJOO','DELL','LG','PRIXTON','JETWING','MEDION','THOMSON','PRIMUX','DENVER','PEAQ','AORUS','BMAX','DEEPGAMING','GIGABYTE','JUMPER','MEGAPORT','RAZER','REALME','VANT']

laptop_name=[]
laptop_brand=[]

for box in boxes:
  time.sleep(5)
  
  try:

    name = box.find_all('p')[0].get_text().strip().replace(' "','"').replace(' GB','GB').replace('Portátil - ','').replace('Portátil  - ','').replace('Portátil ','')
    name=re.sub(', Windows.?1.*|, W.?1.*|..Sin sistema operativo|Blanco|Negro|Gris|Plata|Gris espacial|..No disponible|..Integrada|espacial|  |..Teclado Magic Keyboard Touch ID','',name)
    
    data = {'Name':name}
    
    laptop_name.append(data)

    print(name)

    for i in brands:
      i_lw = i.lower()
      if i_lw in name.lower():
        print(i)

        data = {'Brand':i}
        
        laptop_brand.append(data)

  except:
    pass
```

###Get Product Price

```python
boxes = soup.find_all('div',attrs={'class':'StyledPriceRow-jah2p3-4 eTxUJZ'})

laptop_price=[]

#GET PRICE

for price in boxes:
  time.sleep(5)
  
  try:
    
    #Check if the product has a discount, so it has another price

    price1 = price.find_all('span')[3].get_text().strip()
    price1 = re.findall("\d+", price1)
    
    #If the product has a discount results that price1 is empty, so we get the price from somewhere else

    if len(price1)==0:
      price2 = price.find_all('span')[1].get_text().strip()
      price2 = re.findall("\d+", price2)
      one_cifre=float('.'.join(price2))

      data = {'Price':one_cifre}
      laptop_price.append(data)

      print(one_cifre)
    else:
      price3 = price.find_all('span')[4].get_text().strip()
      price3 = re.findall("\d+", price3)
      one_cifre=float('.'.join(price3))

      data = {'Price':one_cifre}
      laptop_price.append(data)

      print(one_cifre)
  except:
    pass

```

###Get Product Link

```python
boxes = soup.find_all('div',attrs={'class':'StyledListItem-sc-7l2z3t-0 bilJsB'})

laptop_link=[]

for links in boxes:
  time.sleep(5)
  
  try:

    a = links.find_all('a')[0]
    link = ('https://www.mediamarkt.es'+a.get('href'))
  
    data = {'Links':link}
    
    laptop_link.append(data)

    print(link)
  
  except:
    pass
```

###Get Reviews Counts

```python
boxes = soup.find_all('div',attrs={'class':'StyledRatingWrapper-sc-15s0par-0 gIJkAn StyledRating-q99jve-1'})

laptop_review_c=[]

for review in boxes:
  time.sleep(5)
  
  try:

    review_count = review.find_all('span')[0].get_text().strip().replace('(','').replace(')','')
    review_count = int(review_count)

    print(review_count)
  
    data = {'Review counts':review_count}
    
    laptop_review_c.append(data)
  
  except:
    pass
```

###Review score

```python
boxes = soup.find_all('div',attrs={'class':'StyledRatingWrapper-sc-15s0par-0 gIJkAn StyledRating-q99jve-1'})

laptop_review_s=[]

for review_s in boxes:
  time.sleep(5)
  
  try:

    review_s = str(review_s)

    full_starts = re.findall('<div class="Icon-sc-1vrq823-0 hhunIT" data-test="mms-fully-rated-star"><svg viewbox="0 0 24 24"><path d="M2.18,9.86A.49.49,0,0,1,2,9.32l.08-.25a.49.49,0,0,1,.44-.35l6.39-.51,2.45-5.89A.51.51,0,0,1,11.87,2h.26a.49.49,0,0,1,.47.32l2.46,5.89,6.39.51a.49.49,0,0,1,.44.35l.08.25a.49.49,0,0,1-.15.54L17,14l1.48,6.21a.51.51,0,0,1-.19.53L18,20.9a.5.5,0,0,1-.56,0L12,17.6,6.53,20.93a.5.5,0,0,1-.56,0l-.22-.15a.51.51,0,0,1-.19-.53L7,14Z"></path></svg></div>',review_s)
    half_stars = re.findall('ratingStarGrad',review_s)
    
    starts= len(full_starts) + (len(half_stars)/4)

    print(starts)

    data = {'Review score':starts}
    
    laptop_review_s.append(data)
  
  except:
    pass
```

###Create the DataFrame

```python
laptop_n_df=pd.DataFrame(laptop_name)
laptop_p_df=pd.DataFrame(laptop_price)
laptop_l_df=pd.DataFrame(laptop_link)
laptop_b_df=pd.DataFrame(laptop_brand)
laptop_rc_df=pd.DataFrame(laptop_review_c)
laptop_rs_df=pd.DataFrame(laptop_review_s)
```

```python
laptop_product_s = []

for row in range(len(laptop_n_df.index)):
    data = {'Product status':'Brand New'}
    laptop_product_s.append(data)

laptop_ps_df=pd.DataFrame(laptop_product_s)
```

```python
df_merged=pd.concat([laptop_n_df,laptop_p_df,laptop_b_df,laptop_rc_df,laptop_rs_df,laptop_ps_df,laptop_l_df], axis=1)
df_merged
```
